```python
%%writefile main.cu
#include <stdio.h>
#include <stdlib.h>

float* read_csv(const char* filename, int num_rows, int num_cols) {
    // Open the file with the filename provided. 'r' opens the file for reading.
    FILE* file = fopen(filename, "r");

    // Allocate memory for 1D Array
    float* data = (float*)malloc(num_rows * num_cols * sizeof(float));

    // Iterate over the number of rows
    for (int row = 0; row < num_rows; ++row) {
        // Iterate over the number of columns
        for (int col = 0; col < num_cols; ++col) {
            // Reads a float value from the file and stores it in the data array at the correct position.
            fscanf(file, "%f,", &data[row * num_cols + col]);
        }
    }

    // Close the file
    fclose(file);

    // Return the pointer to the array
    return data;
}

// The main function - program execution starts here
int main() {
    // The path and filename of the CSV data file
    const char* filename = "Y_train.csv";

    // Specifies the number of rows and columns to read from the CSV file
    int num_rows = 10;  // Number of rows to read
    int num_cols = 10;  // Number of columns per row

    // Calls the read_csv function and expects back an array filled with data
    float* data = read_csv(filename, num_rows, num_cols);

    // Loop over each element in the data array and print its value
    for (int i = 0; i < num_rows; ++i) {
        for (int j = 0; j < num_cols; ++j) {
            // Print each float value as an integer after rounding to the nearest whole number
            printf("%d ", (int)(data[i * num_cols + j] + 0.5f));
        }
        // After printing all columns of a row, output a newline character to separate rows
        printf("\n");
    }

    // After finishing using the data, free the memory allocated for the data array
    // to prevent memory leaks
    free(data);
    // Exit the main function
    return 0;
}
```

```python
!nvcc main.cu -o main
```

```python
!./main
```
